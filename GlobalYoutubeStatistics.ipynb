```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
url = 'Global YouTube Statistics.csv'
data = pd.read_csv(url, encoding = 'unicode_escape')
data.describe()
```

```python
data.columns
```

```python
data.dtypes
```

```python
y = data['video views']
x = data['Unemployment rate']
```

```python
sns.catplot(x = 'category', data = data, kind = 'count')
plt.xticks(rotation = 90)
plt.show()
```

```python
data.isna().sum()
```

```python
data.shape
```

```python
columns = ['rank', 'Youtuber', 'category', 'Country' ]
only_few_cols = data[columns]
only_few_cols
```

```python
print(only_few_cols.shape)
print(only_few_cols.isna().sum())
```

```python
means_by_country = data.pivot_table(values = 'rank', index = 'Country')
df_country = pd.DataFrame(means_by_country)
x = df_country.index
y = df_country['rank']
sns.catplot(x = x, y = y, kind = 'point')
plt.xticks(rotation = 90)
plt.show()
```

```python
mean_income = data.loc[:, ('lowest_monthly_earnings', 'rank')]
plt.scatter(mean_income['lowest_monthly_earnings'], mean_income['rank'], c = mean_income['lowest_monthly_earnings'])
plt.show()
```
